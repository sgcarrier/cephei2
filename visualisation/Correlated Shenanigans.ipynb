# Correlated test and how to determine the true number of fines

## How about we just start with 1 coarse?

```python
delaySteps = 1 #ps
jitterRMSofFirstFine = 5 #ps
#jitterSteps = 0.1 #ps

length = 10
codePsLengths = [5]*length
```

So we have a perfect delayed source and the TDC is jittered in its "perceived" code. The codes are also of perfect equal length. Truly, we are blessed

```python
import numpy as np

lengthOfTime= 50 #ps
CORR_DATA = np.zeros((lengthOfTime//delaySteps, length))
for t in range(0, lengthOfTime, delaySteps):
    # Normal distribution with jitter = sigma
    data = np.random.normal(t, jitterRMSofFirstFine, 1000)
    # Remove negative values
    data = data[data >= 0] 
    # Generate simulated code conversion we would get
    bins = np.arange(0, (length+1)*10, 10)
    CORR_DATA[t], bounds = np.histogram( data, bins)
    

    
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for t in range(0, lengthOfTime, 1):
    zs = np.arange(0, lengthOfTime, delaySteps)
    ys = CORR_DATA[t]
    xs = (np.arange(0, len(bounds)-1))
    ax.bar(xs, ys, zs=t, zdir='y', color=np.random.rand(3,), alpha=0.8)

ax.set_xlabel('Code')
ax.set_ylabel('Time')
ax.set_zlabel('Count')

plt.show()
```
